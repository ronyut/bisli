```python

def optimize_f1(x):
    p = np.empty_like(label)
    option_dict = { k: t(v) for k, t, v in zip(
        ['n_estimators','base_score','learning_rate','reg_lambda','reg_alpha'],
        [lambda x: int(max(x,0)),
         lambda x: round(min(max(x,0.01),0.99),2),
         lambda x: round(max(x,0),2),
         lambda x: round(max(x,0),2),
         lambda x: round(max(x,0),2)
        ],
        x
    )}
    option_dict['random_state']=1234
    option_dict['booster']='dart'
    for i in range(data.shape[0]-1):
        gb = xgb.XGBClassifier(**option_dict)
        train_x = np.vstack([data[0:i,:],data[i+2:,:]])
        train_y = np.hstack([label[:i],label[i+2:]])
        gb.fit(train_x,train_y)
        p[i] = gb.predict(data[i:i+1,:])
        

        precision = sum(np.logical_and(p,label)) / sum(p)
        recall = sum(np.logical_and(p,label)) / sum(label)
        f1 = 2*precision*recall/(precision+recall)
    option_dict['precision']=round(precision,4)
    option_dict['recall']=round(recall,4)
    option_dict['f1']=round(f1,4)
    log.append(option_dict)
    print(option_dict)
    return -f1
```
